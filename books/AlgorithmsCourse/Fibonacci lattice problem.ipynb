```python
import math
import sys

n = 50
phi = (math.sqrt(5.0) + 1.0) / 2.0
phi_inv = phi - 1.0
ga = 2.0 * phi_inv * math.pi

points = []

for i in range(-n, n + 1):
    longitude = ga * i
    longitude = (longitude % phi) - phi if longitude < 0 else longitude % phi
    latitude = math.asin(2.0 * float(i) / (2.0 * n + 1.0))
    points.append((latitude, longitude))
    
#     print("{}-th point: ".format(i + n + 1))
#     print("\tLongitude is {}".format(longitude))
#     print("\tLatitude is {}".format(latitude))

# print(points)
print ([p[0] for p in points])
print ([p[1] for p in points])

p0 = [p[0] for p in points]
p1 = [p[1] for p in points]
print ("same when p0 sorted = {}".format(p0 == sorted(p0)))
print ("same when p1 sorted = {}".format(p1 == sorted(p1)))


```
