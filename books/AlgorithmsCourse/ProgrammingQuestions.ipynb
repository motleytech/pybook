```python
"""
hw 1-1
"""

def mergeAndCountInv(data, start, num):
    "merge the subparts and counts inversions"
    invs = 0
    inData = data[start : start + num]
    l1 = num//2
    i, j = 0, l1
    offset = start
    while i < l1 and j < num:
        if inData[i] <= inData[j]:
            data[offset] = inData[i]
            i += 1
        else:
#             invs += (j - (offset - start))
            invs += (l1 - i)
            data[offset] = inData[j]
            j += 1
        offset += 1

    if i < l1:
        data[offset:start+num] = inData[i:l1]
    else:
        data[offset:start+num] = inData[j:num]

    return invs


def sortAndCountInv(arr, start, num):
    "sort the data and count the inversions"
    if num == 1:
        return 0

    x = sortAndCountInv(arr, start, num//2)
    y = sortAndCountInv(arr, start + num//2, num - num//2)
    z = mergeAndCountInv(arr, start, num)

    return x + y + z


def countInversions(data):
    "wrapper for counting inversions"
    res1 = sortAndCountInv(data, 0, len(data))
    return (res1)


def getData():
    'return the input array'
    with open('IntegerArray.txt') as fl:
        data = fl.readlines()

    data = [int(d.strip()) for d in data]
    return data


def solveProblem():
    "solve the hw problem"
    data = getData()
    # print (data[:10])
    inversions = countInversions(data)
    print ("Number of inversions: {}".format(inversions))

solveProblem()


```

```python
"""
hw 1-2
"""

def median3(p1, p2, p3):
    'return median of 3 numbers'
    if p1 > p2:
        # p1 p2
        if p3 > p1:
            return p1
        # p1 p2,p3
        if p2 > p3:
            return p2
        return p3
    # p2 p1
    if p3 > p2:
        return p2
    # p2 p1,p3
    if p1 > p3:
        return p1
    return p3


def classPartition(data, l, r, ppos=0):
    'partition method from the class notes'
    if ppos == 0:
        p, pIndex = data[l], l
    if ppos == 1:
        p, pIndex = data[r-1], (r-1)
    if ppos == 2:
        numels = r - l
        if numels % 2:
            # odd
            mid = l + (numels // 2)
        else:
            mid = l + (numels // 2) - 1

        p1, p2, p3 = data[l], data[r-1], data[mid]
        p = median3(p1, p2, p3)
        if p == p1:
            pIndex = l
        elif p == p2:
            pIndex = r-1
        else:
            pIndex = mid

    data[pIndex], data[l] = data[l], data[pIndex]
    i = l + 1
    for j in range(l+1, r):
        if data[j] < p:
            data[i], data[j] = data[j], data[i]
            i += 1
    data[l], data[i-1] = data[i-1], data[l]
    return i-1

def quickSortInner(data, start, num, ppos):
    'sorter'
    comps = num - 1
    pIndex = classPartition(data, start, start+num, ppos)

    # recurse on both parts
    num1 = pIndex - start
    if num1 > 1:
        comps += quickSortInner(data, start, num1, ppos)
    num2 = num - num1 - 1
    if num2 > 1:
        comps += quickSortInner(data, pIndex + 1, num2, ppos)

    return comps

def quickSort(data, ppos=0):
    'wrapper'
    return quickSortInner(data, 0, len(data), ppos)


def getData():
    'load data from input file'
    with open('QuickSort.txt') as fl:
        data = fl.readlines()

    data = [int(d.strip()) for d in data]
    return data

def solve():
    'solve'
    # import random
    # data = range(30)
    # random.shuffle(data)
    data = getData()
    result = sorted(data)
    # print (data[:10])
    for ppos in range(3):
        data0 = data[:]
        comps = quickSort(data0, ppos=ppos)
        if data0 != result:
            print ("Error in set: {}".format(ppos))
            return
        print (data0[:10], comps)

solve()
```
