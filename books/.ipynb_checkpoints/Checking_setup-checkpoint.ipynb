```python
def fibo(n):
    if n <= 2:
        return 1
    a, b = 1, 1
    while n > 2:
        n -= 1
        a, b = b, a+b
    return b

print [fibo(k) for k in xrange(1,10)]
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 3*np.pi, 500)
plt.title('A simple chirp');

plt.plot(x, np.sin(x**2))


```
